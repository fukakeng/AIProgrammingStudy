```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 課題1
## 2変数正規分布に従う乱数をプロットする
- そのままプロット
- 標準化してプロット
    - サンプルからのμ, Σを使用して計算
    - パラメータからのμ, Σを使用して計算
- 無相関化してプロット

```python
def standardize(data, mean=None, cov=None):
    if mean is None:
        mean = data.mean(axis=0)

    if cov is None:
        cov = np.cov(data.T)

    return (data - mean) / np.sqrt(np.diag(cov))


def decorrelate(data, cov=None):
    if cov is None:
        cov = np.cov(data.T)

    la, eig_s = np.linalg.eig(cov)
    return np.dot(eig_s.T, data.T).T


def plot_data(data):
    print(f'平均:\n{data.mean(axis=0)}\n')
    print(f'共分散行列:\n{np.cov(data.T)}')
    sns.distplot(data[:, 0])
    sns.distplot(data[:, 1])
    sns.jointplot(data[:, 0], data[:, 1], s=15)
```

```python
# x = np.random.rand(100, 2)
# mean = a.mean(axis=0)
# sigma = np.cov(a.T)
mean = np.array([3, 10])
sigma = np.array([[15, 3], [3, 4]])
data = np.random.multivariate_normal(mean, sigma, 5000)
```

```python
print('そのままプロット')
plt.figure()
plot_data(data)

print('標準化(サンプルからのμ, Σ)')
plt.figure()
standardized_data = standardize(data)
plot_data(standardized_data)

print('標準化(パラメータからのμ, Σ)')
plt.figure()
standardized_data = standardize(data, mean, sigma)
plot_data(standardized_data)

print('無相関化')
plt.figure()
decorrelated_data = decorrelate(data)
plot_data(decorrelated_data)
```

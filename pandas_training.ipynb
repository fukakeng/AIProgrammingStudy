## 欠損値

- 実務だと小数点以下のレベルの精度差であれば削ってしまってもいい

```python
#欠損割合を表示
(df.isnull().sum()/df.isnull().count()*100).sort_values(ascending = False)[:20]
```

## dataframeの結合
- mergeでSQLのjoinのような形で結合できる

```python
ccb = pd.read_csv('../home-credit-default-risk/credit_card_balance.csv')
ccb.head()
```

```python
'''
SQLならば

SELECT *
FROM
    tbl
INNER JOIN
    ccb
ON tbl.SK_ID_CURR = ccb.SK_ID_CURR



LEFT OUTER JOINする場合は以下のようにhow='left'にする。

df.merge(ccb, on='SK_ID_CURR', how='left')
'''
merge_df = df.merge(ccb, on='SK_ID_CURR', how='inner')

print(merge_df.shape)
merge_df.head()
```

## 集約
- SQLでいうgroup by

```python
'''
SQLならdfをtblとすると

SELECT 
    FLAG_OWN_CAR, CODE_GENDER, AVG(AMT_ANNUITY)
FROM
    tbl
GROUP BY
    FLAG_OWN_CAR, CODE_GENDER
    
と等価
'''

df.groupby(['FLAG_OWN_CAR', 'CODE_GENDER']).agg({'AMT_CREDIT': 'mean'})
```

- 集約のキーがindexになるため、reset_indexで戻すことが多い

```python
df.groupby(['FLAG_OWN_CAR', 'CODE_GENDER']).agg({'AMT_CREDIT': 'mean'}).reset_index()
```

- numpyを組み合わせた集約

```python
import numpy as np

df.groupby(['CODE_GENDER']).agg({'AMT_CREDIT': ['mean', lambda x: np.percentile(x, q=20)]}).reset_index()
```

## shift関数
- SQLのlag関数に相当
- 例としてはあるユーザーの前月の売上を当月レコードに新たなカラムとして追加するなど

```python
groupby = ccb.query('SK_ID_CURR == 378907 | SK_ID_CURR == 363914') \
    .groupby('SK_ID_CURR') \
    .apply(lambda x: x.sort_values(by='MONTHS_BALANCE', ascending=False))
groupby
```

```python
groupby['AMT_BALANCE_SHIFTED'] = groupby['AMT_BALANCE'].shift(periods=-1)
groupby[['AMT_BALANCE', 'AMT_BALANCE_SHIFTED']]
```

## pivot_table
- クロス集計
- index:縦軸, columns:横軸, values:集計する値
- 集計の値はデフォルトでは平均 aggfuncで指定可能

```python
# mean
pd.pivot_table(df, index='CODE_GENDER', columns='NAME_CONTRACT_TYPE', values='AMT_CREDIT', aggfunc=np.mean)
```

```python
# count
pd.pivot_table(df, index='CODE_GENDER', columns='NAME_CONTRACT_TYPE', values='AMT_CREDIT', aggfunc='count')
```

問7

```python
bureau = pd.read_csv('../home-credit-default-risk/bureau.csv')
```

```python
days_credits = bureau[['SK_ID_CURR', 'SK_ID_BUREAU', 'DAYS_CREDIT']]
target_id_currs = [162297, 215354]
result_df = pd.DataFrame(columns=['SK_ID_CURR', 'SK_ID_BUREAU', 'DAYS_CREDIT', 'PREVIOUS_DAYS_DIFF'])

for id in target_id_currs:
    extracted_days_credits = days_credits.query('SK_ID_CURR == @id').sort_values('DAYS_CREDIT', ascending=False)
    extracted_days_credits['PREVIOUS_DAYS_DIFF'] = extracted_days_credits['DAYS_CREDIT'] - extracted_days_credits['DAYS_CREDIT'].shift(periods=-1)
    result_df = pd.concat([result_df, extracted_days_credits], axis=0)

result_df.index = range(len(result_df))
result_df.fillna(0)
```
